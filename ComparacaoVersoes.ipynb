```
import pandas as pd

dados_fixed = pd.DataFrame(data = {'number_terms' : [10,15,20,25,30,35,40], 
                                   'time' : [14,16,16,20,21,25,26]})
dados_cached = pd.DataFrame(data = {'number_terms' : [10,15,20,25,30,35,40],
                                    'time' : [12,14,18,16,19,18,21]})
dados_work_stealing = pd.DataFrame(data = {'number_terms' : [10,15,20,25,30,35,40],
                                           'time' : [12,11,14,14,15,17,22]})
```

```
import matplotlib.pyplot as plt   
import seaborn as sns
sns.set(rc={'figure.figsize':(14,10)})
```

```
sns.lineplot(x="number_terms", y="time", data=dados_fixed)
sns.lineplot(x="number_terms", y="time", data=dados_cached)
sns.lineplot(x="number_terms", y="time", data=dados_work_stealing)
plt.xlabel('Quantidade de termos', fontsize=15)
plt.ylabel('Tempo de execução (milissegundos)', fontsize=15)
plt.legend(['Fixed','Cached','Work stealing'])
```
